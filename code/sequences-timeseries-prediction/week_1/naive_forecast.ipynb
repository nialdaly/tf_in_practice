### Timeseries Fundamentals
A `timeseries` is an ordered sequence of values that are usually equally spaced over time.

`Univariate` timeseries are those where a single value exists at each timestep.

`Multivariate` timeseries are those where multiple values exist at each timestep.

`Trend` - when a timeseries moves in a specific direction.

`Seasonality` - when patterns in a timeseries tend to repeat at predictable intervals.


`Autocorrelation` - timeseries where no trend or seasonality exists. The timeseries correlates with a delayed copy of itself. This delayed copy is known as a `lag`.

https://otexts.com/fpp2/stationarity.html


`Naive Forecast` - taking timeseries data and assuming that the next value will be the same as the previous value.

Naive forecasting can be a useful tool when we need to get a baseline.

`Fixed partitioning`
The process of splitting a timeseries into 3 periods:
- `Training Period`

- `Validation Period`

- `Test Period`

If the timeseries contains seasonality then we want to ensure that each period contains a whole number of seasons.


`Roll-forward Partitioning`
By starting with a short training period and gradually increasing it by a defined time interval (1 day/week).

At each iteration, the model is trained on the training period and the model is used to predict the next day/week into the validation period.


### Metrics
`errors = forecasts - actual`

`mse = np.square(errors).mean()`

`rmse = np.sqrt(mse)`

`mae = np.abs(errors).mean()`

`mape = np.abs(errors / x_valid).mean()`

If large errors are potentially dangerous and cost alot more than small errors, then the `mse` maybe a better evaluation metric. 

### Moving Average & Differencing
`Moving average` removes alot of noise from the timeseries.

`Differencing` - process of removing trend and seasonality from a timeseries.

`Smoothing` - process of removing past noise from a timeseries.
